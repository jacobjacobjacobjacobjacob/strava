```python
def filter_by_period(
    df: pd.DataFrame,
    start_date: str = None,
    end_date: str = None,
    year: int = None,
    month: int = None,
) -> pd.DataFrame:
    """
    Filter DataFrame by time parameters.

    :param
        df (pd.DataFrame): The DataFrame containing Strava data.
        start_date (str, optional): Start date in 'YYYY-MM-DD' format.
        end_date (str, optional): End date in 'YYYY-MM-DD' format.
        year (int, optional): Year to filter by.
        month (int, optional): Month to filter by.

    :returns
        pd.DataFrame: Filtered DataFrame.
    """

    filtered_df = df.copy()

    # Sort by date
    filtered_df = filtered_df.sort_values(by="date", ascending=True)

    if start_date is not None:
        filtered_df = filtered_df[filtered_df["date"] >= start_date]
    if end_date is not None:
        filtered_df = filtered_df[filtered_df["date"] <= end_date]
    if year is not None:
        filtered_df = filtered_df[filtered_df["date"].dt.year == year]
    if month is not None:
        filtered_df = filtered_df[filtered_df["date"].dt.month == month]
    return filtered_df
```

```python
def rename_columns(df: pd.DataFrame) -> pd.DataFrame:
    """
    Rename columns for consistency.

    :param
        df (pd.DataFrame): The DataFrame containing Strava data.

    :returns
        pd.DataFrame: DataFrame with renamed columns.
    """

    filtered_df = df.copy()
    if "gear_id" in filtered_df.columns:
        filtered_df.rename(columns={"gear_id": "ride_type"}, inplace=True)
    if "start_date" in filtered_df.columns:
        filtered_df.rename(columns={"start_date": "date"}, inplace=True)
    if "moving_time" in filtered_df.columns:
        filtered_df.rename(columns={"moving_time": "duration"}, inplace=True)
    if "total_elevation_gain" in filtered_df.columns:
        filtered_df.rename(
            columns={"total_elevation_gain": "elevation_gain"}, inplace=True
        )

    # Convert all text in all rows to lowercase
    for column in filtered_df.select_dtypes(include=["object"]).columns:
        # Ensure the column contains string values
        if filtered_df[column].apply(lambda x: isinstance(x, str)).all():
            filtered_df[column] = filtered_df[column].str.lower()
        
    return filtered_df
```

```python
def convert_units(df: pd.DataFrame) -> pd.DataFrame:
    """
    Convert units such as distance and speed.

    :param
        df (pd.DataFrame): The DataFrame containing Strava data.

    :returns
        pd.DataFrame: DataFrame with converted units.
    """
    filtered_df = df.copy()

    if "distance" in filtered_df.columns:
        filtered_df["distance"] = filtered_df["distance"].apply(m_to_km)
    if "duration" in filtered_df.columns:
        filtered_df["duration"] = filtered_df["duration"].apply(sec_to_h)
    if "average_speed" in filtered_df.columns:
        filtered_df["average_speed"] = filtered_df["average_speed"].apply(ms_to_kph)
    if "max_speed" in filtered_df.columns:
        filtered_df["max_speed"] = filtered_df["max_speed"].apply(ms_to_kph)

    if "date" in filtered_df.columns:
        filtered_df["date"] = pd.to_datetime(filtered_df["date"])

    return filtered_df
```

```python
def m_to_km(m: float) -> float:
    """
    Convert meters to kilometers.

    :param
        m (float): Distance in meters.

    :return
        float: Distance in kilometers.
    """
    return m / 1000


def ms_to_kph(ms: float) -> float:
    """
    Convert meters per second to kilometers per hour.

    :param
        ms (float): Speed in meters per second.

    :return
        float: Speed in kilometers per hour.
    """
    return ms * 3.6


def sec_to_h(sec: float) -> float:
    """
    Convert seconds to hours.

    :param
        sec (float): Duration in seconds.

    :return
        float: Duration in hours.
    """
    return sec / 3600
```

```python
def filter_columns(cleaned_df: pd.DataFrame) -> tuple:
    """
    Split the cleaned DataFrame into outdoor_df and indoor_df based on ride_type.

    :param
        cleaned_df (pd.DataFrame): The DataFrame containing cleaned Strava data.

    :returns
        tuple: DataFrames for all activities, outdoor activities, and indoor activities.
    """

    columns = {
        "all": [
            "id",
            "date",
            "month",
            "day_of_week",
            "start_time",
            "end_time",
            "duration",
            "distance",
            "elevation_gain",
            "average_speed",
            "max_speed",
            "average_heartrate",
            "max_heartrate",
            "suffer_score",
            "suffer_score_bucket",
            "elevation_ratio",
            "sport_type"
        ],


    }

    filtered_df = cleaned_df[columns["all"]].copy()

    return filtered_df
```

```python
def add_month_column(df: pd.DataFrame) -> pd.DataFrame:
    """
    Add month abbreviation to DataFrame.

    :param
        df (pd.DataFrame): The DataFrame containing Strava data.

    :returns
        pd.DataFrame: DataFrame with month abbreviation added.
    """
    df["month"] = df["date"].dt.month_name().str.slice(stop=3).str.lower()
    return df
```

```python
def add_day_of_week(df: pd.DataFrame) -> pd.DataFrame:
    """
    Add day of week to DataFrame.

    :param
        df (pd.DataFrame): The DataFrame containing Strava data.

    :returns
        pd.DataFrame: DataFrame with day of week added.
    """
    weekdays = ["mon", "tue", "wed", "thu", "fri", "sat", "sun"]
    df["day_of_week"] = df["date"].dt.weekday.map(lambda x: weekdays[x])
    return df
```

```python
def add_time_columns(df: pd.DataFrame) -> pd.DataFrame:
    """
    Add start_time and end_time columns based on duration.

    :param
        df (pd.DataFrame): The DataFrame containing Strava data.

    :returns
        pd.DataFrame: DataFrame with added time columns.
    """
    filtered_df = df.copy()

    # Check if date column is timezone-aware
    if filtered_df["date"].dt.tz is None:
        # Localize to UTC
        filtered_df["date"] = filtered_df["date"].dt.tz_localize("UTC")

    # Convert to local timezone
    filtered_df["date"] = filtered_df["date"].dt.tz_convert(local_tz)

    # Calculate end_time based on duration
    filtered_df["end_time"] = filtered_df["date"] + pd.to_timedelta(
        filtered_df["duration"], unit="h"
    )

    filtered_df["start_time"] = filtered_df["date"].dt.strftime("%H:%M")
    filtered_df["end_time"] = filtered_df["end_time"].dt.strftime("%H:%M")

    return filtered_df
```

```python
def add_elevation_rate(df: pd.DataFrame) -> pd.DataFrame:
    """
    Add elevation gained per kilometer.

    :param
        df (pd.DataFrame): The DataFrame containing Strava data.

    :returns
        pd.DataFrame: DataFrame with elevation ratio added.
    """
    df["elevation_ratio"] = df.apply(
        lambda row: (row["distance"] / row["elevation_gain"]) if row["elevation_gain"] != 0 else 0,
        axis=1
    )
    return df
```

```python
def sort_and_reset_index(df: pd.DataFrame) -> pd.DataFrame:
    """
    Sort DataFrame by 'date' column and reset index.

    :param
        df (pd.DataFrame): The DataFrame containing Strava data.

    :returns
        pd.DataFrame: Sorted and reindexed DataFrame.
    """
    df_sorted = df.copy()
    df_sorted = df.sort_values(by="date")

    df_sorted.reset_index(drop=True, inplace=True)
    df_sorted.index += 1  # Shift index to start from 1

    return df_sorted
```

```python
def convert_date_to_yyyymmdd(df: pd.DataFrame) -> pd.DataFrame:
    """
    Convert 'date' column to format YYYY-MM-DD.

    :param
        df (pd.DataFrame): DataFrame with a 'date' column.

    :return
        pd.DataFrame: DataFrame with 'date' column converted to YYYY-MM-DD format.
    """
    df_cleaned = df.copy()
    df_cleaned["date"] = df_cleaned["date"].dt.date
    return df_cleaned
```

```python
def map_ride_type(df: pd.DataFrame, gear_dict) -> pd.DataFrame:
    """
    Map ride_type values based on gear dictionary and keep sport_type as "Run" unchanged.

    :param
        df (pd.DataFrame): The DataFrame containing Strava data.
        gear_dict (dict): Dictionary to map ride types.

    :returns
        pd.DataFrame: DataFrame with mapped ride types.
    """
    if "sport_type" in df.columns and "ride_type" in df.columns:
        df.loc[df["sport_type"] != "run", "sport_type"] = df["ride_type"].map(gear_dict)
    return df
```

```python
def replace_nan_values(df: pd.DataFrame) -> pd.DataFrame:
    """
    Replace NaN values with the mean for columns with missing heart rate data.

    :param
        df (pd.DataFrame): The DataFrame containing Strava data.

    :returns
        pd.DataFrame: DataFrame with NaN values replaced.
    """
    columns_to_fill = ["max_heartrate", "average_heartrate", "suffer_score"]

    for column in columns_to_fill:
        if column in df.columns:
            mean_value = df[column].mean()
            df[column] = df[column].fillna(mean_value)

    return df
```

```python
def add_suffer_score_buckets(df):
    labels = ["very low", 'low', 'medium', 'high', "very high"]
    df['suffer_score_bucket'] = pd.qcut(df['suffer_score'],q=5, labels=labels)
    return df
    
```

```python
def clean_df(df):
    cleaned_df = (df.pipe(rename_columns)
          .pipe(convert_units)
          .pipe(filter_by_period, year=2024)
          .pipe(add_month_column)
          .pipe(add_day_of_week)
          .pipe(add_time_columns)
          .pipe(add_elevation_rate)
          .pipe(add_suffer_score_buckets)
          .pipe(map_ride_type, gear)
          .pipe(convert_date_to_yyyymmdd)
          .pipe(replace_nan_values)
          .pipe(filter_columns)
          .pipe(sort_and_reset_index)

    )
    return cleaned_df
```
